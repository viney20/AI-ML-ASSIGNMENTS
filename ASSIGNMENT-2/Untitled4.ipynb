```python
#Question 1

import numpy as np
x=np.random.rand(10,1)
print("Formed Matrix Is \n",x)

print("Mean Of The Above Matrix Is =",np.sum(x)/len(x))


#Second Method
print("Mean Of The Above Matrix Is =",np.mean(x))
```

```python
#Question 2

x=np.random.rand(20,1)
print("Formed Matrix Is \n",x)

print("Variance of the formed matrix is =",np.var(x))


print("Standard deviation of the formed matrix is =",np.std(x))
#Second Method For standard deviation
print("Standard deviation of the formed matrix is =",np.sqrt(np.var(x)))
```

```python
#Question 3

A=np.random.rand(10,20)
B=np.random.rand(20,25)

x=np.matmul(A,B)
print("Matrix Formed After Multiplication\n",x)


print("Sum Of Newly Formed Matrix = ",np.sum(x))
```

```python
#Question 4
#f(x)=1 / (1 + exp(-x)) 
from math import exp
y=np.random.rand(10,1)
print(y)
l=[]
for i in y:
    l.append(1/(1+exp(-(i))))
    
print("\n\n\nAfter performing given function")
print(np.array(l).reshape(10,1))
```
