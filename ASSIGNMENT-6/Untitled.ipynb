```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data=pd.read_csv("data.csv")
data=data.dropna()
data.head()
```

```python
data.shape
data.describe(include='all')
```

```python
data.columns
```

```python
j=0
for i in data.columns:
    if(data[i].dtypes!=object):
        plt.figure(j)
        sns.distplot(data[i],rug=True)
        j+=1
```

```python
sns.scatterplot(data=data,x='MinTemp',y='MaxTemp')
```

```python
with sns.axes_style('white'):
    sns.jointplot(x='Evaporation',y='Rainfall',kind='hex',data=data,color='k')
```

```python
sns.pairplot(data)
```
