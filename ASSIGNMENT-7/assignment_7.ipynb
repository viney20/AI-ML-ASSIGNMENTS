```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
```

```python
# Import Dataset and write description about it
df=pd.read_csv('titanic.csv')
df.head()
#Description-
#This is a data from The Titanic
#This data shows the name,age,sex,fare they paid and whether they survived or not
```

```python
print(df.shape)
df.dropna()
print(df.shape)
# find count of missing data
#There Is No Missing Data
```

```python
#Univariate Analysis - Plot the univariate graphs and find the outliers.
#To find outliers in univariate analysis boxplot can be used
plt.figure(0)
sns.boxplot(x=df['Fare'])
plt.figure(1)
sns.boxplot(x=df['Age'])
```

```python
#Bivariate Analysis - Plot the bivariate graphs and find the relationship between the variables. Explain your findings and 
#inference of the graphs.
df['WithSomeone']=df['Siblings/Spouses Aboard']+df['Parents/Children Aboard']

plt.figure(0)
sns.swarmplot(data=df,x="Pclass",y='Fare') 
#this shows variation in fares in different class.From the data we get to observe that high variation is in 1st class only
plt.figure(1)
sns.swarmplot(data=df,x="Sex",y='Fare')
#this shows affect of sex on fare and we get to know that there is not effect of sex on fare of ship
plt.figure(2)
sns.scatterplot(data=df,x='Age',y='Fare')
#this shows average fare according to age group and we get to know that mostly fare was less than 100
plt.figure(3)
sns.catplot('Sex',data=df,kind='count')
#this shows more male than female
plt.figure(4)
sns.catplot('Pclass',data=df,kind='count')
#this shows more passenger in class3
plt.figure(5)
sns.catplot('Survived',data=df,kind='count')
#this shows the ratio of passengers that survived
plt.figure(6)
sns.catplot('WithSomeone',data=df,kind='count')
#this shows whether passenger was travelling alone or not



```

```python
#Multivariate Analysis - Plot the multivariate graphs and find the relationship between the variables. Explain your findings and
#inference of the graphs. 
plt.figure(0)
sns.catplot('Pclass',data=df,hue='Sex',kind='count')
#this shows variation in sex in different classes
```
